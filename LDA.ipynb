```python
import glob
import codecs
import nltk
import re

import pprint
from nltk.corpus import stopwords

from many_stop_words import get_stop_words
nltk.download('stopwords')
```

```python
from __future__ import absolute_import, division, print_function
```

```python
file_path = sorted (glob.glob("./corona.txt"))
```

```python
nltk_words = list(stopwords.words('english'))
manystopwords = list(get_stop_words('en'))
```

```python
import nltk.data
nltk.download('punkt')
```

```python
def process(file):
    string = u''
    with codecs.open(file, "r", "utf-8") as book_file:
        string += book_file.read()
    tokenizer = nltk.data.load('nltk:tokenizers/punkt/english.pickle')
    
    
    sentences = tokenizer.tokenize(string)
    
    nltk_words = list(stopwords.words('english'))
    manystopwords = list(get_stop_words('en')) 
    def sentence_to_wordlist(raw):
        clean = re.sub("[^a-zA-Z]"," ", raw)
        words = clean.split()
        lowercase = [word.strip().lower() for word in words]
        filtered_word_list = lowercase[:]  # make a copy of the word_list
        
        for word in lowercase:  # iterate over word_list
            if word in manystopwords: 
                filtered_word_list.remove(word) #
        return filtered_word_list
    
    words = []
    for sentence in sentences:
        if len(sentence) > 0:
            words.append(sentence_to_wordlist(sentence))
    return words
```

```python
words = process(file_path[0])
```

```python
flat_list = [item for sublist in words for item in sublist]
```

```python
word = []
```

```python
for file in file_path:
    word.append ( [item.lower() for sublist in process(file) for item in sublist] )
```

```python
from gensim import corpora
```

```python
dictionary = corpora.Dictionary(word)
dictionary.save('./corona.dict')  
print(dictionary)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in word]
```

```python
import gensim.models.ldamodel as lda
```

```python
mymodel = lda.LdaModel(corpus, id2word=dictionary, num_topics=30)
```
